```python
Dataset_raw = np.array(Dataset_df)
Dataset = np.random.permutation(Dataset_raw)

X_train = Dataset[:12000,:9]
Y_train = Dataset[:12000,-3:]
#Y_train[:,1] = (Y_train[:,1] - 0.99) * 10000
X_test  = Dataset[12000:,:9]
Y_test  = Dataset[12000:,-3:]
#Y_test[:,1] = (Y_test[:,1] - 0.99) * 10000

'''
X_train = Dataset[:12000,:9]
Y_train = Dataset[:12000,-3:]
X_test  = Dataset[12000:,:9]
Y_test  = Dataset[12000:,-3:]
'''

scaler_X_train = sp.StandardScaler().fit(X_train)
X_train_scl = scaler_X_train.transform(X_train)
scaler_X_test = sp.StandardScaler().fit(X_test)
X_test_scl = scaler_X_test.transform(X_test)
scaler_Y_train = sp.StandardScaler().fit(Y_train)
Y_train_scl = scaler_Y_train.transform(Y_train)
scaler_Y_test = sp.StandardScaler().fit(Y_test)
Y_test_scl = scaler_Y_test.transform(Y_test)

X_train_scl_input = X_train_scl
X_train_scl_output = X_train_scl
```

```python
fig, ax = plt.subplots(6,3,figsize=(20,15))
for i in range(9):
    plt.subplot(6,3,i+1)
    plt.hist(X_train[:,i] , bins = 50)
    plt.subplot(6,3,i+10)
    plt.hist(X_train_scl[:,i] , bins = 50)
```

```python
encoding_dim = 3
input_data = tf.keras.layers.Input(shape = (9,))

encoder = tf.keras.layers.Dense(8  , activation = 'relu')(input_data)
encoder = tf.keras.layers.Dense(16 , activation = 'relu')(encoder)
encoder = tf.keras.layers.Dense(32 , activation = 'relu')(encoder)
encoder = tf.keras.layers.Dense(16 , activation = 'relu')(encoder)
encoder = tf.keras.layers.Dense(8  , activation = 'relu')(encoder)
encoder_output = tf.keras.layers.Dense(encoding_dim)(encoder)

decoder = tf.keras.layers.Dense(8  , activation = 'relu')(encoder_output)
decoder = tf.keras.layers.Dense(16 , activation = 'relu')(decoder)
decoder = tf.keras.layers.Dense(32 , activation = 'relu')(decoder)
decoder = tf.keras.layers.Dense(16 , activation = 'relu')(decoder)
decoder = tf.keras.layers.Dense(8  , activation = 'relu')(decoder)
decoder = tf.keras.layers.Dense(9)(decoder)

autoencoder = tf.keras.models.Model(inputs = input_data , outputs = decoder)
autoencoder.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),
                    loss = 'mse',
                    metrics = ['accuracy'])
autoencoder.summary()
history = autoencoder.fit(X_train_scl_input , X_train_scl_output , epochs = 500 , batch_size = 1000 , validation_split = 0.2 , verbose = 2)

'''
encoder = tf.keras.models.Sequential([
        tf.keras.layers.Dense(units = 16  , activation = 'relu' , input_dim = 9) ,
        tf.keras.layers.Dense(units = 64  , activation = 'relu'),
        tf.keras.layers.Dense(units = 128 , activation = 'relu'),
        tf.keras.layers.Dense(units = 256 , activation = 'relu'),
        tf.keras.layers.Dense(units = 128 , activation = 'relu'),
        tf.keras.layers.Dense(units = 64  , activation = 'relu'),
        tf.keras.layers.Dense(units = 16  , activation = 'relu'),
        tf.keras.layers.Dense(units = 3                        )
        ])

decoder = tf.keras.models.Sequential([
        tf.keras.layers.Dense(units = 16  , activation = 'relu' , input_dim = 3) ,
        tf.keras.layers.Dense(units = 64  , activation = 'relu'),
        tf.keras.layers.Dense(units = 128 , activation = 'relu'),
        tf.keras.layers.Dense(units = 256 , activation = 'relu'),
        tf.keras.layers.Dense(units = 128 , activation = 'relu'),
        tf.keras.layers.Dense(units = 64  , activation = 'relu'),
        tf.keras.layers.Dense(units = 16  , activation = 'relu'),
        tf.keras.layers.Dense(units = 16                        )
        ])

model = tf.keras.models(inputs = encoder , outputs = decoder)
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),
              loss = 'binary_crossentropy',
              metrics = ['accuracy','mse'])
model.summary()
history = model.fit(X_train_scl , Y_train_scl , epochs = 500 , batch_size = 1000 , validation_split = 0.2 , verbose = 1)
'''
```

```python
score = autoencoder.evaluate(X_test_scl , X_test_scl , batch_size = 1000 , verbose = 0)
print('score:',score)
print('Accuracy:',score[1]*100,'%')
```
